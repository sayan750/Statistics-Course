### 1. 

**The difference between t-test & z-test are:-**

1. T-test refers to a type of parametric test that is applied to identify, how the means of two sets of data differ from one another when variance is not given. On the other hand, Z-test implies a hypothesis test which ascertains if the means of two datasets are different from each other when variance is given.

2. T-Test based on Student-t distribution while Z-Test based on Normal distribution.

3. T-Test population variance is unknown. On the other hand Z-Test population variance is known.

4. T-Test sample size is small. Z-Test sample size is large.

**T-test:- Suppose I am conducting a study to determine if there is a significant difference in the mean scores of two groups on a certain variable, such as test scores. I have a sample of 30 students in each group and I want to use a t-test to determine if there is a statistically significant difference in the means of the two groups. I would use a t-test because my sample size is relatively small (less than 50), and the variance of the two groups may be different.**

**Z-test:- Suppose I am analyzing a large dataset of 10,000 customers and I want to determine if there is a significant difference between the proportion of customers who buy a particular product from two different marketing campaigns. I use a z-test to compare the two proportions. In this case, a z-test is appropriate because the sample size is large (more than 30), and you are comparing proportions rather than means.**

### 3.

**A Type 1 error occurs when the null hypothesis is rejected, even though it is true. In other words, the test concludes that there is a significant effect or difference, when in reality there is none. This error is also known as a "false positive."An example scenario for a Type 1 error could be a medical trial testing a new drug. The null hypothesis is that the drug has no effect on the patients. However, due to chance, the results of the trial show a significant improvement in the treatment group compared to the control group. If the researchers conclude that the drug is effective and reject the null hypothesis, even though it is true, they have made a Type 1 error.**

**A Type 2 error occurs when the null hypothesis is not rejected, even though it is false. In other words, the test fails to detect a significant effect or difference, when in reality there is one. This error is also known as a "false negative." An example scenario for a Type 2 error could be a medical test for a rare disease. The null hypothesis is that the patient does not have the disease. However, due to the test's low sensitivity, the patient receives a negative result, even though they do have the disease. If the doctors conclude that the patient does not have the disease and fail to reject the null hypothesis, even though it is false, they have made a Type 2 error.**

**Bayes's theorem is a mathematical formula that helps calculate the probability of an event occurring, given some prior knowledge or evidence. It is named after the 18th-century English statistician and philosopher, Thomas Bayes. The theorem states that the probability of an event A given some evidence B can be calculated as follows:-**

P(A|B) = P(B|A) * P(A) / P(B)

Where:

P(A|B) is the probability of event A given evidence B.

P(B|A) is the probability of evidence B given event A.

P(A) is the prior probability of event A.

P(B) is the prior probability of evidence B.

**Here's an example to illustrate Bayes's theorem:- Suppose a medical test is used to diagnose a certain disease, which affects 1% of the population. The test has a 90% accuracy rate, meaning it correctly identifies 90% of people with the disease and 10% of people without it. If a person tests positive for the disease, what is the probability that they actually have it?**

Let's use Bayes's theorem to find out:

P(A) = 0.01 (the prior probability of having the disease)

P(B|A) = 0.90 (the probability of testing positive given that the person has the disease)

P(B|~A) = 0.10 (the probability of testing positive given that the person does not have the disease)

Using these values, we can calculate the probability of actually having the disease given a positive test result:

P(A|B) = P(B|A) * P(A) / P(B)

P(A|B) = 0.90 * 0.01 / [(0.90 * 0.01) + (0.10 * 0.99)]

P(A|B) = 0.08 or 8%

So, even though the test has a high accuracy rate, there is still only an 8% chance that a person who tests positive for the disease actually has it. This highlights the importance of considering both the accuracy of the test and the prior probability of the disease when interpreting test results.

**A confidence interval is a range of values that is likely to contain the true value of a population parameter, such as the mean or proportion. It is a statistical tool used to estimate the level of uncertainty associated with a sample estimate, based on the variability observed in the sample data.**

To calculate a confidence interval, you need to know the sample size, sample mean or proportion, standard deviation or standard error of the sample, and the level of confidence desired.

Here's an example:

Suppose you want to estimate the average height of all college students in the United States, and you take a random sample of 100 students. You measure their heights and find that the sample mean height is 68 inches, and the standard deviation of heights in the sample is 3 inches.

To calculate the 95% confidence interval for the population mean height, you would use the following formula:

Confidence interval = sample mean ± (critical value x standard error)

The critical value depends on the level of confidence desired and the sample size. For a 95% confidence interval and a sample size of 100, the critical value is approximately 1.96.

The standard error is the standard deviation of the sample divided by the square root of the sample size. In this case, the standard error is 3 / sqrt(100) = 0.3 inches.

Plugging in the values, we get:

Confidence interval = 68 ± (1.96 x 0.3) = (67.41, 68.59)

Therefore, we can be 95% confident that the true population mean height of college students in the United States falls within the range of 67.41 to 68.59 inches.

**Here is a sample problem that demonstrates how to use Bayes' Theorem:-**

Suppose that a certain disease affects 1 in 1,000 people. A new diagnostic test for the disease has been developed that is 99% accurate, meaning that if a person has the disease, there is a 99% chance that the test will correctly identify them as positive, and if they do not have the disease, there is a 1% chance that the test will incorrectly identify them as positive. If a person tests positive for the disease, what is the probability that they actually have the disease?

Solution:

Let's use Bayes' Theorem to calculate the probability that a person has the disease given that they have tested positive:

P(Disease|Positive) = P(Positive|Disease) * P(Disease) / P(Positive)

We know that:

P(Positive|Disease) = 0.99 (the test correctly identifies 99% of people with the disease as positive)

P(Disease) = 0.001 (1 in 1,000 people have the disease)

To calculate P(Positive), we need to consider both the true positives (people who have the disease and test positive) and the false positives (people who do not have the disease but test positive):

P(Positive) = P(Positive|Disease) * P(Disease) + P(Positive|No disease) * P(No disease)

= 0.99 * 0.001 + 0.01 * 0.999

= 0.01098

So we can now calculate the probability that a person has the disease given that they have tested positive:

P(Disease|Positive) = P(Positive|Disease) * P(Disease) / P(Positive)

= 0.99 * 0.001 / 0.01098

= 0.0902 (or approximately 9.02%)

Therefore, even though a person tests positive for the disease, there is only a 9.02% chance that they actually have the disease.

**The margin of error in a confidence interval is a measure of how much uncertainty there is in the estimated population parameter. It represents the range within which the true value of the population parameter is expected to lie, given a certain level of confidence. The margin of error is typically expressed as a plus or minus value around the point estimate of the parameter.**

Sample size affects the margin of error in two ways. First, as the sample size increases, the standard error of the mean decreases, leading to a smaller margin of error. This is because larger samples provide more information about the population, and therefore the estimate of the population parameter is likely to be more accurate. Second, larger sample sizes also allow for a higher level of confidence, which leads to a larger value of Z* and a larger margin of error.

**For example, suppose we want to estimate the average height of a certain population with 95% confidence, and we take a sample of 50 individuals. If the sample mean height is 68 inches, and the standard deviation of the population is known to be 3 inches, then the margin of error can be calculated as:-**

Margin of error = 1.96 * (3 / sqrt(50)) ≈ 0.84 inches

This means we can be 95% confident that the true average height of the population lies within the range of 67.16 to 68.84 inches.

If we were to increase the sample size to 100 individuals, while keeping all other factors the same, the margin of error would decrease:

Margin of error = 1.96 * (3 / sqrt(100)) ≈ 0.60 inches

This means we can be even more confident that the true average height of the population lies within the narrower range of 67.4 to 68.6 inches.

In general, larger sample sizes lead to smaller margins of error and more precise estimates of population parameters. However, collecting larger samples can be more time-consuming and costly, so researchers must balance the benefits of smaller margins of error with the practical constraints of their study.

z = (x - μ) / σ

where x is the data point, μ is the population mean, and σ is the population standard deviation.

Plugging in the values, we get:

z = (75 - 70) / 5
z = 1

This means that the data point of 75 is one standard deviation above the population mean. The z-score tells us how many standard deviations a data point is from the population mean. A positive z-score means that the data point is above the population mean, while a negative z-score means that the data point is below the population mean.

In this case, a z-score of 1 indicates that the data point is higher than the mean by one standard deviation. This can also be interpreted as the data point being in the 84th percentile of the population distribution, meaning that 84% of the population falls below this value.

Alternative Hypothesis: The weight loss drug is significantly effective (i.e., the population mean weight loss is different from zero). We can use a t-test to test this hypothesis. we can calculate the t-statistic:-

t = (6 - μ) / (2.5 / √50) = 10√2(6 - μ) / 5

Assuming a 95% confidence level, we need to find the critical t-value for a two-tailed test with 49 degrees of freedom (df=n-1). We can find this value using a t-distribution table or calculator. For a two-tailed test and 49 degrees of freedom, the critical t-value is approximately ±2.009.

If the absolute value of the t-statistic is greater than the critical t-value, we reject the null hypothesis and conclude that the weight loss drug is significantly effective. Otherwise, we fail to reject the null hypothesis.

Let's assume that the null hypothesis is true (i.e., μ = 0). Then, the t-statistic becomes:

t = (6 - 0) / (2.5 / √50) = 12√2

Since the absolute value of the t-statistic (12√2) is greater than the critical t-value (2.009), we reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level.

**To conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance, we can use a two-sample t-test.**

First, we define the null and alternative hypotheses:

Null hypothesis: The two teaching methods have no significant difference in student performance.
Alternative hypothesis: The two teaching methods have a significant difference in student performance.

We can set the significance level to 0.01, which means that we are willing to accept a 1% chance of making a Type I error (rejecting the null hypothesis when it is actually true).

To perform the two-sample t-test, we first calculate the pooled standard deviation:

sp = sqrt(((n1-1)*s1^2 + (n2-1)*s2^2) / (n1 + n2 - 2))

where n1 and n2 are the sample sizes, s1 and s2 are the sample standard deviations, and sp is the pooled standard deviation.

For our example, we have:

n1 = n2 = 30 (assuming equal sample sizes)
s1 = 6
s2 = 5

sp = sqrt(((30-1)*6^2 + (30-1)*5^2) / (30 + 30 - 2)) = 5.52

Next, we calculate the t-statistic:

t = (x1 - x2) / (sp * sqrt(2/n))

where x1 and x2 are the sample means, n is the sample size, and sp is the pooled standard deviation.

For our example, we have:

x1 = 85
x2 = 82
n = 30

t = (85 - 82) / (5.52 * sqrt(2/30)) = 3.06

Using a t-table with 58 degrees of freedom (df = n1 + n2 - 2), we can find the critical t-value for a two-tailed test with a significance level of 0.01. The critical t-value is approximately ±2.660.

Since our calculated t-value (3.06) is greater than the critical t-value (2.660), we reject the null hypothesis and conclude that there is a significant difference in student performance between the two teaching methods.
