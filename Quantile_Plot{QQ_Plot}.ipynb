**QQ-Plot**

A QQ plot, short for Quantile-Quantile plot, is a graphical tool used in statistics to assess whether two datasets come from the same distribution or not.

To create a QQ plot, we first sort the data points in each dataset from smallest to largest. Then, we plot the corresponding quantiles of one dataset against the quantiles of the other dataset. If the two datasets come from the same distribution, the points on the QQ plot should lie on a straight line.

If the points deviate from the straight line, it suggests that the two datasets have different distributions. The degree of deviation from the straight line can also provide information on how different the distributions are.

QQ plots are particularly useful for comparing small datasets or datasets where the distribution is not known or is difficult to model. They are commonly used in fields such as finance, engineering, and environmental science.

```python
import scipy.stats as stat
import pylab
import numpy as np
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot

def plot_dat(df,feature):
  plt.figure(figsize=(10,8))
  plt.subplot(1,2,1)
  sns.histplot(df[feature], kde = True)
  plt.subplot(1,2,2)
  stat.probplot(df[feature], dist = 'norm', plot = pylab)
  plt.show()
```

```python
plot_dat(df,'sepal_width')
```

```python
plot_dat(df,'sepal_length')
```

```python
plot_dat(df,'petal_width')
```
